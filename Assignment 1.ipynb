Data clean up – correcting the data types

a. Which all variables need to be brought to numeric types?

Answer)Variables Reviews,Installs,Price should be brought up as numeric 


```python
read_data.dtypes
```

Price variable – remove $ sign and convert to float

Installs – remove ‘,’ and ‘+’ sign, convert to integer

Convert all other identified columns to numeric

```python
def remove_sign(variable,sign):
    read_data[variable].replace(to_replace=sign,value="",regex=True,inplace=True)

def convert_datatype(variable,datatype):
    read_data[variable]=read_data[variable].astype(datatype)

remove_sign("Installs",["\,","\+"])
remove_sign("Price",["\$"])
convert_datatype("Installs","int")
convert_datatype("Price","float")
convert_datatype("Reviews","int")

read_data.dtypes
```
